# **Project Summary -**

This project focuses on analyzing digital payment data from the PhonePe Pulse dataset to understand transaction behavior, user engagement, and insurance adoption trends across India. The primary goal is to derive meaningful insights that can help in strategic business decision-making.

The dataset consists of three major components: transaction data, user data, and insurance data. Transaction data includes information such as transaction count, amount, type, state, and time period. User data provides insights into registered users and app opens, helping measure engagement levels. Insurance data captures the adoption and growth of insurance-related transactions across regions.

The project begins with data cleaning and preprocessing, where missing values and duplicates were handled, and data types were standardized. A new feature called "engagement" was created using app opens and registered users to better understand user activity.

Exploratory Data Analysis (EDA) was conducted using various visualizations such as bar charts, line charts, histograms, scatter plots, heatmaps, and pair plots. These visualizations helped uncover key trends, including the dominance of certain states in transaction volume, steady growth in digital payments over time, and varying levels of user engagement across regions.

The analysis also highlighted that while a few states contribute significantly to overall transactions, many regions show lower adoption, indicating strong potential for growth. User data revealed increasing engagement and app usage, while insurance data showed gradual growth but limited penetration, suggesting opportunities for expansion.

Hypothesis testing was performed to validate relationships between variables such as transaction count and amount, as well as differences in engagement across states. Statistical methods like correlation tests and ANOVA were used to support findings.

Overall, this project provides a comprehensive understanding of digital payment trends, user behavior, and emerging business opportunities. The insights derived can help PhonePe optimize its strategies, target underperforming regions, and expand its services effectively.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H0 (Null Hypothesis): There is no relationship between transaction count and transaction amount.

H1 (Alternate Hypothesis): There is a significant relationship between transaction count and amount.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import pearsonr

corr, p_value = pearsonr(agg_txn["count"], agg_txn["amount"])

print("Correlation:", corr)
print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

Test Used: Pearson Correlation Test

##### Why did you choose the specific statistical test?

Reason: Used to measure relationship between two numerical variables.

Conclusion: Since p-value < 0.05, we reject the null hypothesis. There is a significant relationship between transaction count and amount.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H0: There is no difference in app engagement across states.

H1: There is a difference in engagement across states.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

states = agg_user["state"].unique()

groups = [agg_user[agg_user["state"] == s]["engagement"] for s in states[:5]]

f_stat, p_value = f_oneway(*groups)

print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

Test Used: ANOVA

##### Why did you choose the specific statistical test?

Reason: Used to compare means across multiple groups.

Conclusion: If p-value < 0.05, engagement differs across states.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H0: Insurance transactions do not vary across years.

H1: Insurance transactions vary across years.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
years = agg_ins["year"].unique()

groups = [agg_ins[agg_ins["year"] == y]["amount"] for y in years]

f_stat, p_value = f_oneway(*groups)

print("P-value:", p_value)
````
